```python
import pyarrow.parquet as pq
from pyarrow import csv
import pandas as pd
```

```python
green_df = pq.read_table(source="./data/green_tripdata_2024-01.parquet", use_threads=True).to_pandas()
yellow_df = pq.read_table(source="./data/yellow_tripdata_2024-01.parquet", use_threads=True).to_pandas()
```

```python
g_set = set(green_df.columns)
y_set = set(yellow_df.columns)
```

```python
g_unique = sorted(list(g_set - y_set))
y_unique = sorted(list(y_set - g_set))
g_y_intersect = sorted(list((g_set & y_set)))
print("g_unique", g_unique)
print("y_unique", y_unique)
print("g_y_intersect", g_y_intersect)
```

```python
label_size = 10
g_sample = green_df.sample(n=10000)
g_sample.hist(xlabelsize=label_size, ylabelsize=label_size, figsize=(15,15))
```

```python
label_size = 10
y_sample = yellow_df.sample(n=10000)
y_sample.hist(xlabelsize=label_size, ylabelsize=label_size, figsize=(15,15))
```
