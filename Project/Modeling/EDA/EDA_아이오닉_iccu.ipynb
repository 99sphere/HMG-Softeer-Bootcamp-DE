```python
def plot_per_day_post_count_with_events(df, title, day_start, day_end, events=None):
    full_date_daily = pd.date_range(start=day_start, end=day_end, freq='D').to_frame(index=False, name='full_range_date')
    full_date_daily['full_range_date'] = full_date_daily['full_range_date'].astype(str)

    # df = df.loc[df["likes"] > per_post_likes_threshold]
    df = df.loc[:, ["created_day", "views"]]
    
    per_day_view = df.groupby(df.created_day).agg({ # 일별 조회수의 합, 게시물의 수
        'views': ['sum', 'count'],  
    })

    per_day_view.columns = per_day_view.columns.droplevel(0)
    # per_day_view = per_day_view.loc[per_day_view['sum'] > per_day_views_threshold]

    full_range_df = pd.merge(full_date_daily, per_day_view, left_on='full_range_date', right_on='created_day', how='left')
    full_range_df = full_range_df.fillna(0)

    fig = go.Figure()        
    fig.add_trace(
        go.Bar(
            x=full_range_df['full_range_date'],
            y=full_range_df['count'],
            name='count'
        )
    )
    
    if events is not None:
        y_max = int(full_range_df['count'].max())
        for event in events:
            date, color, text = event
            if day_start <= date <= day_end:
                fig.add_shape(type="line", x0=date, x1=date, y0=0, y1=y_max, line=dict(color=color, width=2, dash='dash'))
                fig.add_trace(go.Scatter(x=[date],y=[-1], text=[text], mode="text"))

    fig.update_layout(
        title=title,
        xaxis_title='date',
        yaxis_title='',
    )

    fig.show()
    return full_range_df
```

```python
def plot_per_day_normalized_score(df, title, day_start, day_end, events=None, smoothing=''):
    full_date_daily = pd.date_range(start=day_start, end=day_end, freq='D').to_frame(index=False, name='full_range_date')
    full_date_daily['full_range_date'] = full_date_daily['full_range_date'].astype(str)

    views_df = df.loc[:, ["created_day", "views"]]    
    per_day_views = views_df.groupby(df.created_day).agg({ # 일별 조회수의 합, 게시물의 수
        'views': ['sum', 'count'],  
    })
    per_day_views.columns = per_day_views.columns.droplevel(0)

    likes_df = df.loc[:, ["created_day", "likes"]]    
    per_day_likes = likes_df.groupby(df.created_day).agg({ # 일별 조회수의 합, 게시물의 수
        'likes': ['sum'],  
    })
    per_day_likes.columns = per_day_likes.columns.droplevel(0)


    cmt_count_df = df.loc[:, ["created_day", "cmt_count"]]    
    per_day_cmt_count = cmt_count_df.groupby(df.created_day).agg({ # 일별 조회수의 합, 게시물의 수
        'cmt_count': ['sum'],  
    })
    per_day_cmt_count.columns = per_day_cmt_count.columns.droplevel(0)

    
    # normalize
    per_day_views = per_day_views.assign(window_sum_count=per_day_views['count'].rolling(30, center=False).sum())
    per_day_views = per_day_views.assign(window_sum_views=per_day_views['sum'].rolling(30, center=False).sum())
    per_day_likes = per_day_likes.assign(window_sum_likes=per_day_likes['sum'].rolling(30, center=False).sum())
    per_day_cmt_count = per_day_cmt_count.assign(window_sum_cmt_count=per_day_cmt_count['sum'].rolling(30, center=False).sum())

    # per_day_views = per_day_views.assign(window_sum_count=per_day_views['count'].sum())
    # per_day_views = per_day_views.assign(window_sum_views=per_day_views['sum'].sum())
    # per_day_likes = per_day_likes.assign(window_sum_likes=per_day_likes['sum'].sum())
    # per_day_cmt_count = per_day_cmt_count.assign(window_sum_cmt_count=per_day_cmt_count['sum'].sum())

    per_day_views = per_day_views.assign(norm_count=per_day_views['count']/per_day_views['window_sum_count'])    
    per_day_views = per_day_views.assign(norm_views=per_day_views['sum']/per_day_views['window_sum_views'])
    per_day_likes = per_day_likes.assign(norm_likes=per_day_likes['sum']/per_day_likes['window_sum_likes'])
    per_day_cmt_count = per_day_cmt_count.assign(norm_cmt_count=per_day_cmt_count['sum']/per_day_cmt_count['window_sum_cmt_count'])

    # join
    full_range_df = pd.merge(full_date_daily, per_day_views, left_on='full_range_date', right_on='created_day', how='left')
    full_range_df = pd.merge(full_range_df, per_day_likes, left_on='full_range_date', right_on='created_day', how='left')
    full_range_df = pd.merge(full_range_df, per_day_cmt_count, left_on='full_range_date', right_on='created_day', how='left')
    
    full_range_df = full_range_df.fillna(0)
    full_range_df = full_range_df.assign(norm_score=full_range_df['norm_count']+full_range_df['norm_views']+full_range_df['norm_likes']+full_range_df['norm_cmt_count'])

    fig = go.Figure()    
    fig.add_trace(
        go.Bar(
            x=full_range_df['full_range_date'],
            y=full_range_df['norm_score'],
            name='norm_score'
        )
    )

    if smoothing=='ewm':
        full_range_df = full_range_df.assign(score_smoothed=full_range_df['norm_count'].ewm(alpha=0.6, adjust=True).mean())
        fig.add_trace(
            go.Scatter(
                x=full_range_df['full_range_date'],
                y=full_range_df['score_smoothed'],
                mode='lines',
                name='count_ewm'
            )
        )
    elif smoothing=='ma':
        full_range_df = full_range_df.assign(view_smoothed=full_range_df['norm_count'].rolling(3, center=False).mean())
        fig.add_trace(
            go.Scatter(
                x=full_range_df['full_range_date'],
                y=full_range_df['score_smoothed'],
                mode='lines',
                name='count_ma'
            )
        )
            
    if events is not None:
        y_max = int(full_range_df['norm_score'].max())
        for event in events:
            date, color, text = event
            if day_start <= date <= day_end:
                fig.add_shape(type="line", x0=date, x1=date, y0=0, y1=y_max, line=dict(color=color, width=2, dash='dash'))
                fig.add_trace(go.Scatter(x=[date],y=[-1], text=[text], mode="text"))

    fig.update_layout(
        title=title,
        xaxis_title="data",
        yaxis_title="",
    )

    fig.show()
    return full_range_df
```

```python
def plot_per_day_post_count_plotly_with_events(df, title, day_start, day_end, events=None, per_post_likes_threshold=-1, per_day_views_threshold=-1, x_axis='date', y_axis='count', smoothing=''):
    full_date_daily = pd.date_range(start=day_start, end=day_end, freq='D').to_frame(index=False, name='full_range_date')
    full_date_daily['full_range_date'] = full_date_daily['full_range_date'].astype(str)

    # df = df.loc[df["likes"] > per_post_likes_threshald]
    df = df.loc[:, ["created_day", "views"]]
    
    per_day_view = df.groupby(df.created_day).agg({ # 일별 조회수의 합, 게시물의 수
        'views': ['sum', 'count'],  
    })

    per_day_view.columns = per_day_view.columns.droplevel(0)
    # per_day_view = per_day_view.loc[per_day_view['sum'] > per_day_views_threshold]

    full_range_df = pd.merge(full_date_daily, per_day_view, left_on='full_range_date', right_on='created_day', how='left')
    full_range_df = full_range_df.fillna(0)

    fig = go.Figure()    
    # smoothing
    if smoothing=='ewm':
        full_range_df = full_range_df.assign(count_smoothed=full_range_df['count'].ewm(alpha=0.6, adjust=True).mean())
        fig.add_trace(
            go.Scatter(
                x=full_range_df['full_range_date'],
                y=full_range_df['count_smoothed'],
                mode='lines',
                name='count_ewm'
            )
        )
    elif smoothing=='ma':
        full_range_df = full_range_df.assign(count_smoothed=full_range_df['count'].rolling(3, center=False).mean())
        fig.add_trace(
            go.Scatter(
                x=full_range_df['full_range_date'],
                y=full_range_df['count_smoothed'],
                mode='lines',
                name='count_ma'
            )
        )
    
    fig.add_trace(
        go.Bar(
            x=full_range_df['full_range_date'],
            y=full_range_df['count'],
            name='count'
        )
    )
    y_max = int(full_range_df['count'].max())
    
    if events is not None:
        for event in events:
            date, color, text = event
            if day_start <= date <= day_end:
                fig.add_shape(type="line", x0=date, x1=date, y0=-0.3, y1=y_max, line=dict(color=color, width=2, dash='dash'))
                fig.add_trace(go.Scatter(x=[date],y=[-1], text=[text], mode="text"))

    fig.update_layout(
        title=title,
        xaxis_title=x_axis,
        yaxis_title=y_axis,
    )

    fig.show()
    return full_range_df
```
